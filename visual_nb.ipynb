```python
import sympy as sp
import numpy as np
from scipy.linalg import expm
from qiskit.visualization import plot_bloch_vector
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define Identity matrix & Pauli matrices
I = sp.eye(2)
sigma_x = sp.Matrix([[0, 1], [1, 0]])
sigma_y = sp.Matrix([[0, -sp.I], [sp.I, 0]])
sigma_z = sp.Matrix([[1, 0], [0, -1]])

# Convert from sympy data type to native float type
def sympy_to_normal(sympy_matrix):
    return [1, float(sp.N(sympy_matrix[1])), float(sp.N(sympy_matrix[2]))]

# Rotate spinor by a given angle around a specified axis
def rotate_spinor(spinor, angle, axis):
    n_vector = [sp.sin(axis[1])*sp.cos(axis[2]), sp.sin(axis[1])*sp.sin(axis[2]), sp.cos(axis[1])]
    spinor_matrix = sp.Matrix([[sp.cos(spinor[1]/2)], [sp.sin(spinor[1]/2)*sp.exp(sp.I*spinor[2])]])
    print("Spinor Matrix: ", spinor_matrix)

    sigma_dot_axis = sigma_x*n_vector[0] + sigma_y*n_vector[1] + sigma_z*n_vector[2]
    print("Sigma Dot Axis: ", sigma_dot_axis)

    # Rotation matrix
    rot_matrix = sp.cos(angle/2)*I - sp.I*sp.sin(angle/2)*sigma_dot_axis
    print("Rotation Matrix: ", rot_matrix)

    # Spinor after rotation
    rotated_spinor = rot_matrix*spinor_matrix
    print("Rotated Spinor: ", rotated_spinor)

    # extract the rotated spinor's  and 
    if sp.re(rotated_spinor[0,0])== 0:
        if sp.im(rotated_spinor[0,0]) >= 0:
            A = sp.pi/2
        else:
            A = -sp.pi/2
        A = sp.pi/2
    else:
        A = sp.atan(sp.im(rotated_spinor[0,0])/sp.re(rotated_spinor[0,0]))

    if sp.re(rotated_spinor[1,0] )== 0:
        if sp.im(rotated_spinor[1,0]) >= 0:
            B = sp.pi/2
        else:
            B = -sp.pi/2
    else:
        B = sp.atan(sp.im(rotated_spinor[1,0])/sp.re(rotated_spinor[1,0]))
    theta = 2*sp.acos(sp.sqrt((sp.re(rotated_spinor[0,0]**2 + (sp.im(rotated_spinor[0,0])**2)))))
    phi = B - A

    return [1, theta, phi]

# Find single equivalent rotation from two rotations
def find_single_rotation(spinor, angle_1, axis_1, angle_2, axis_2):
    # finding unit vector defining the two axes of rotation
    n1_vect = sp.Matrix([sp.sin(axis_1[1]*sp.cos(axis_1[2])), sp.sin(axis_1[1]*sp.sin(axis_1[2])), sp.cos(axis_1[1])])
    n2_vect = sp.Matrix([sp.sin(axis_2[1]*sp.cos(axis_2[2])), sp.sin(axis_2[1]*sp.sin(axis_2[2])), sp.cos(axis_2[1])])

    # finding the real component of the rotation matrix (cosine term)
    cosine_term = sp.cos(angle_1/2)*sp.cos(angle_2/2) - sp.sin(angle_1/2)*sp.sin(angle_2/2)*n1_vect.dot(n2_vect)
    single_angle_of_rotation = 2*sp.acos(cosine_term)

    # finding the imaginary component of the rotation matrix (sine term)
    sine_term = sp.sin(angle_1/2)*sp.cos(angle_2/2)*n1_vect + sp.cos(angle_1/2)*sp.sin(angle_2/2)*n2_vect - sp.sin(angle_1/2)*sp.sin(angle_2/2)*n1_vect.cross(n2_vect)

    # case when angle of rotation is 0, making sine term 0
    if sine_term != 0:
        axis_of_rotation = sine_term/sp.sin(single_angle_of_rotation/2)
    else:
        axis_of_rotation = [0,0,0]

    # finding cartesian coordinates of the axis of rotation
    rot_theta = sp.acos(axis_of_rotation[2])
    rot_phi = sp.atan(axis_of_rotation[1]/axis_of_rotation[0])

    spinor_after_one_rotation = rotate_spinor(spinor, single_angle_of_rotation, [1, rot_theta, rot_phi])

    return [1, rot_theta, rot_phi], spinor_after_one_rotation

# Define initial spinor and rotation axes
spinor = [1,0,0]
axis1 = [1,sp.pi/2,0]
axis2 = [1,sp.pi/2,0]

# Rotate spinors
rotated_spinor_1 = rotate_spinor(spinor, sp.pi / 2, axis1)
rotated_spinor_2 = rotate_spinor(rotated_spinor_1, sp.pi / 2, axis2)
single_rot_axis, single_rot_spinor = find_single_rotation(spinor, sp.pi/2, axis1, sp.pi/2, axis2)
```
